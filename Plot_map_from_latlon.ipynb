```python
from convertbng.util import convert_bng, convert_lonlat
import pandas as pd
import numpy as np
import folium
df=pd.read_csv("/content/filenames.csv",header=None)
nameoffiles=df.values.tolist()
x=[]
y=[]
for i in range (len(nameoffiles)):
  coordinate_str=nameoffiles[i][0].split('loc')[1].split(".png")[0].split(",")
  coordinate_float=[[float(coordinate_str[0])],[float(coordinate_str[1])]]
  x.append(coordinate_float[0])
  y.append(coordinate_float[1])
x_arr=np.array(x).reshape(-1)
y_arr=np.array(y).reshape(-1)
res_list=convert_lonlat(x_arr,y_arr)
latlong=np.array(res_list)
latlong_=np.transpose(latlong)
latlong_[:,[0, 1]] = latlong_[:,[1, 0]]
mapit = folium.Map( location=[52.667989, -1.464582], zoom_start=6 )
for coord in latlong_:
    folium.Marker( location=[ coord[0], coord[1] ], ).add_to( mapit )

mapit.save('map___.html')
```
